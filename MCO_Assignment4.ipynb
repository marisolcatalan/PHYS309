```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import constants
plt.rcParams.update({"font.size":14, "axes.labelsize":16, "font.family":"sans-serif", 
                     "font.sans-serif":"Arial"})
```

```python
star_types= {"O5":54000, "B5":15200, "A5":8310, "F5":6700,"G5":5660,"K5":4400, "M5":3200}
lm = []
temp_values = []

for i in star_types:
    T = star_types[i] 
    lambda_max = ((2.898E-3)*(1E9))/ T  #Units: nm
    lm.append(lambda_max)
    temp_values.append(T)

plt.scatter(temp_values, lm, marker = "o", color = "cornflowerblue")

plt.text(temp_values[0]-1000, lm[0] +25 , "O5", color= "steelblue")
plt.text(temp_values[1]-700, lm[1] + 50 , "B5", color= "dimgrey")
plt.text(temp_values[2]+200, lm[2] + 25 , "A5", color= "mediumseagreen")
plt.text(temp_values[3]-3250, lm[3] - 25 , "F5", color= "olivedrab")
plt.text(temp_values[4]+750, lm[4] , "G5", color= "darkorange")
plt.text(temp_values[5]+750, lm[5] , "K5", color= "peru")
plt.text(temp_values[6]+800, lm[6]-15, "M5", color= "indianred")

plt.xlabel("Temperature (K)")  
plt.ylabel("Maximum Wavelength (nm)")
plt.title("Temp vs. Max Wavelength Star Types")
```

A note about units: recall from PHYS 208/245 that 1 Volt = 1 Joule/Coulomb. Electrons and protons give us a fundamental unit of electric charge, or elementary charge, $|q| = e = 1.602 \times 10^{−19}$ C. We can also calculate a fundamental unit of energy by calculating the change in kinetic energy of the elementary charge $e$ accelerated through an electric potential of 1 Volt. This new unit of energy is called the electron-Volt or eV. 1 eV = $1.602 \times 10^{−19}$ J.
